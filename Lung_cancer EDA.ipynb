```python
z = df['AGE'].value_counts()
z
```

```python
# line chart

plt.figure(figsize = (10, 6))

sns.lineplot(data = z, marker = 'o', color = 'purple')

plt.title('Age trend analysis')
plt.xlabel('Age')
plt.ylabel('Person counts')
plt.grid()

plt.show()
```

```python
# insights >> highest counts come from 50-60 range
```

```python
# Lung Cancer by gender
sns.countplot(x = 'LUNG_CANCER', hue = 'GENDER', data = df)
```

```python
# insight >> from this data we can see more females are affected from cancer than men but men are less affected 
```

```python
# Univariate analysis

plt.figure(figsize = (10,6))
sns.histplot(df['AGE'].dropna(), kde = True, bins = 20, color = 'skyblue')
plt.title('dist of age')
plt.xlabel('Age')
plt.ylabel('frequency')
plt.show()
```

```python
# Age is noramlly ditributed here
```

```python
# stacked bar Chart

lungcancer_chestpain = pd.crosstab(df['LUNG_CANCER'], df['CHEST_PAIN'])
lungcancer_chestpain
```

```python
lungcancer_chestpain.plot(kind = 'bar', stacked = True)
```
