```python
# now we need to do data encoding as machine enable to understand the categorical data.

from sklearn.preprocessing import OneHotEncoder
onehot = OneHotEncoder()

arrays = onehot.fit_transform(df[["sex"]]).toarray()
df1 = pd.DataFrame(arrays, columns=onehot.get_feature_names_out())
df1
```

```python
df.drop("sex",axis = 1, inplace = True)
df = pd.concat([df,df1],axis = 1)
```

```python
onehot1 = OneHotEncoder()
arrays1 = onehot1.fit_transform(df[["federal_district"]]).toarray()
df2 = pd.DataFrame(arrays1, columns = onehot1.get_feature_names_out())
df2
```

```python
df.drop("federal_district",axis = 1, inplace = True)
df = pd.concat([df,df2], axis = 1)
df
```

```python
city_col = ["federal_district_central","federal_district_volga","federal_district_siberian","federal_district_southern","federal_district_ural","federal_district_far eastern","federal_district_northwestern","federal_district_north caucasian"]
```

```python
for col in city_col:
    plt.figure(figsize=(8,5))
    sns.countplot(x=col,data = df,palette = "coolwarm")
    plt.title(f"Univariate analysis of {col}")
    plt.show()
    
    
# as we can say that central region has the highest vote percentage while the far_eastern region has the lowest percentage in the sample.
```

```python
reas = df["candidate"].value_counts()
plt.pie(reas,labels=reas.index, autopct = "%1.1f%%")

# Most of the people vote to putin while followed the people who struggle to answer. 
```

```python
print(df.groupby(["candidate"])["sex_male","sex_female"].value_counts())

# this shows that the most of the women vote the Putin while most of the men also voted tyhe Putin.
# Slutskiy gets the lowest percentage of vote in the sample.
```

```python
sns.countplot(data=df, x = "income")

# this shows that the middle class of the people vote largest in the election.
# while people who don't reveal their real income vote the lowest.
```

```python
edu = df["education"].value_counts()
plt.pie(edu,labels=edu.index, autopct = "%1.1f%%")

# people who attend the school votes the highest and followed by college while the people who don't go to the school vote the lowest.
```

```python
corr = df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr,annot= True)

# this shows that there is no multicollinearity between the independent variables.
```
